```julia
plot(dates, US_data, xticks=dates[1:5:end], xrotation=45, leg=:topleft, 
    label="US data", m=:o)

xlabel!("date")
ylabel!("confirmed cases in US")
title!("US confirmed COVID-19 cases")

# annotate!(20, US_data[end], text("US", :blue, :left))
```

```julia
plot(dates, US_data, xticks=dates[1:5:end], xrotation=45, leg=:topleft, 
    label="US data", m=:o,
    yscale=:log10)

xlabel!("date")
ylabel!("confirmed cases in US")
title!("US confirmed COVID-19 cases")

# annotate!(20, US_data[end], text("US", :blue, :left))
```

```julia
num_days = length(names(data)[5:end])
```

```julia
numcols = names(data, findall(x -> eltype(x) <: Number, eachcol(data)))
new_data = combine(groupby(data, ["Country/Region"]), numcols .=> sum .=> numcols)
```

```julia
p = plot()
for i in 1:length(countries)
    country_data_row = new_data[new_data[:,"Country/Region"] .== countries[i],:]
    plot!(dates, Vector(country_data_row[1,2:end]))
end
```

```julia
p
```

```julia
p2 = plot()
for i in 1:length(countries)
    country_data_row = new_data[new_data[:,"Country/Region"] .== countries[i],:]
    y_out = Vector(country_data_row[1,2:end])
    y_out[y_out .== 0] .= 1
    plot!(dates, y_out, yscale=:log10)
end
```

```julia
p2
```

```julia
p3 = plot(fmt = :png)
@manipulate for day in 1:num_days
    for i in 1:length(countries)
        country_data_row = new_data[new_data[:,"Country/Region"] .== countries[i],:]
        y_out = Vector(country_data_row[1,2:end])
        y_out[y_out .== 0] .= 1
        date2 = dates[1:num_days]
        y_out2 = y_out[1:num_days]
        plot!(date2, y_out2, yscale=:log10, xlim = (0, num_days))
        p3
    end
end
```
